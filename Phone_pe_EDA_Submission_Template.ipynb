## ***1. Know Your Data***

### Import Libraries

````python
# Import Libraries

import pandas as pd
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
# Load Dataset

data_list = []

files = [f for f in os.listdir() if f.endswith('.json')]

for i, file in enumerate(files):
    with open(file) as f:
        data = json.load(f)

        users = data['data']['usersByDevice']
        total_users = data['data']['aggregated']['registeredUsers']
        app_opens = data['data']['aggregated']['appOpens']

        for row in users:
            data_list.append({
                "Quarter": f"Q{i+1}",
                "Brand": row['brand'],
                "Count": row['count'],
                "Percentage": row['percentage'],
                "Total_Users": total_users,
                "App_Opens": app_opens
            })

df = pd.DataFrame(data_list)
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import json
import pandas as pd
import os

data_list = []

files = [f for f in os.listdir() if f.endswith('.json')]

for i, file in enumerate(files):
    with open(file) as f:
        data = json.load(f)

        users = data['data']['usersByDevice']
        total_users = data['data']['aggregated']['registeredUsers']
        app_opens = data['data']['aggregated']['appOpens']

        for row in users:
            data_list.append({
                "Quarter": f"Q{i+1}",
                "Brand": row['brand'],
                "Count": row['count'],
                "Percentage": row['percentage'],
                "Total_Users": total_users,
                "App_Opens": app_opens
            })

df = pd.DataFrame(data_list)

# Check nulls
df.isnull().sum()

# Remove duplicates (if any)
df.drop_duplicates(inplace=True)
````

### What all manipulations have you done and insights you found?

The dataset was created by extracting data from multiple JSON files and combining them into a single structured DataFrame. Key variables such as brand, user count, percentage share, total users, and app opens were extracted.

Data cleaning steps included checking for missing values and duplicates. No major missing values were found, and duplicates were removed to ensure data consistency. A new column called "Quarter" was created to represent time progression across datasets.

The data was then structured in a tabular format to enable efficient analysis and visualization.
