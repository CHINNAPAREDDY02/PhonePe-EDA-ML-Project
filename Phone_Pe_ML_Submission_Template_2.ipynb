## ***1. Know Your Data***

### Import Libraries

````python
# Import Libraries

import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
# Load Dataset

data_list = []

files = [f for f in os.listdir() if f.endswith('.json')]

for i, file in enumerate(files):
    with open(file) as f:
        data = json.load(f)

        users = data['data']['usersByDevice']
        total_users = data['data']['aggregated']['registeredUsers']
        app_opens = data['data']['aggregated']['appOpens']

        for row in users:
            data_list.append({
                "Quarter": f"Q{i+1}",
                "Brand": row['brand'],
                "Count": row['count'],
                "Percentage": row['percentage'],
                "Total_Users": total_users,
                "App_Opens": app_opens
            })

df = pd.DataFrame(data_list)
````

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

There is no significant difference in user counts across different mobile brands.
There is a significant difference in user counts across different mobile brands.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

groups = [group["Count"].values for name, group in df.groupby("Brand")]
f_stat, p_val = f_oneway(*groups)

print("P-value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

ANOVA (Analysis of Variance)


##### Why did you choose the specific statistical test?

ANOVA is used because we are comparing the means of more than two groups (multiple brands). It helps determine whether at least one brand has significantly different user counts compared to others.


### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

There is no relationship between total users and app opens.
There is a significant relationship between total users and app opens.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

corr, p_val = pearsonr(df["Total_Users"], df["App_Opens"])

print("Correlation:", corr)
print("P-value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test


##### Why did you choose the specific statistical test?

Pearson correlation is used to measure the strength and direction of the linear relationship between two numerical variables (Total Users and App Opens).


### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

There is no significant difference in total users between Quarter 1 and Quarter 4.
There is a significant difference in total users between Quarter 1 and Quarter 4.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

q1 = df[df["Quarter"] == "Q1"]["Total_Users"]
q4 = df[df["Quarter"] == "Q4"]["Total_Users"]

t_stat, p_val = ttest_ind(q1, q4)

print("P-value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

Independent T-Test

##### Why did you choose the specific statistical test?

The T-test is used to compare the means of two groups (Quarter 1 and Quarter 4) to determine if there is a statistically significant difference between them.
